```python
from bs4 import BeautifulSoup
from selenium import webdriver
from time import sleep
from time import time
from random import randint
from IPython.core.display import clear_output
import pandas as pd
import numpy as np
import pickle
import psycopg2
import requests
import os
import re
import ast
```

```python
headers = {'User-Agent': 'Mozilla/5.0',
           'From': 'itchyandscratchy@gmail.com'
          }

data_path = '/Users/amanda/Documents/Projects/insight/data/'
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_hobbies'

hobbylinks = []
# Main summary page
with requests.get(url, timeout=5, headers=headers) as u:     

    soup = BeautifulSoup(u.content)
    for link in soup.find_all('a'):
        hobbylinks.append(link.get('href'))
        
links = [s for s in hobbylinks if s]
linkends = [s for s in links if '/wiki/' in s]
```

```python
linkends
```

```python
base_url = 'https://en.wikipedia.org'

hobbies_df = pd.DataFrame({'hobby':[],
                           'description':[]})

for link in linkends[5:468]:
    try:
        url = base_url + link
        with requests.get(url, timeout=5, headers=headers) as u:     
            soup = BeautifulSoup(u.content)
            #print(soup.prettify())
            title = soup.find('title')
            hobby_title = title.text.replace(' - Wikipedia','').strip()
            print(hobby_title)

            # Main website content
            a = soup.find('div', attrs={'class':'mw-parser-output'})
            desc = a.findAll('p')
            desc_list = [s.text.strip() for s in desc]
            desc_list = [s for s in desc_list if s]
            if len(desc_list) > 1:
                description = ''.join(desc_list[0:2])
            else:
                description = ''.join(desc_list)

            hobbies_df = hobbies_df.append({'hobby':hobby_title,
                                            'description':description}, ignore_index=True)
    except:
        continue
            
```

```python
hobbies = {'Acting & performance':'Acting',
           'Backpacking and hiking':'Backpacking (hiking)',
           'Baking':'Baking',
           'Birdwatching':'Birdwatching',
           'Blogging':'Blog',
           'Board/tabletop games':'Board game',
           'Computer programming':'Computer programming',
           'Cooking':'Cooking',
           'Creative writing':'Creative writing',
           'Dance':'Dance',
           'Language learning':'Second-language acquisition',
           'Knitting or fibre-crafts':'Knitting',
           'Painting and visual art':'Painting',
           'Photography':'Photography',
           'Music':'Music',
           'Gaming':'Video game',
           'Woodworking':'Woodworking',
           'Writing':'Writing',
           'Running':'Running',
           'Cycling':'Cycling',
           'Reading':'Reading'}
```
